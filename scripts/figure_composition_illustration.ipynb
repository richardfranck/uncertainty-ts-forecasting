```python
"""
Motifs and Compositions Illustration

"""
import re
import warnings
warnings.showwarning = lambda m, *a, **k: print(re.sub(r"/Users/[^/]+/", "/Users/USER/", str(m)))

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scienceplots


plt.style.use("science")
sns.set_context("poster", font_scale=1.0)
```

```python
def create_sine_curve(n_points=400):
    """Create x and y coordinates for a sine curve.
    
    Args:
        n_points: Number of points to generate along the curve.
    
    Returns:
        Tuple of (x, y) numpy arrays representing the sine curve.
    """
    x = np.linspace(0, 2 * np.pi, n_points)
    y = np.sin(x)
    return x, y


def get_transition_points():
    """Get transition points where the sine curve changes motif.
    
    Returns:
        List of x-coordinates where motifs transition (minima, maxima, endpoints).
    """
    return [0, np.pi / 2, 3 * np.pi / 2, 2 * np.pi]


def plot_motifs_composition(x, y, transition_points, figsize=(10, 5)):
    """Plot sine curve with motif annotations and transition points.
    
    Visualizes how a continuous function can be decomposed into a sequence
    of motifs (increasing, decreasing) separated by transition points.
    
    Args:
        x: numpy array of x-coordinates.
        y: numpy array of y-coordinates.
        transition_points: List of x-coordinates marking motif transitions.
        figsize: Tuple specifying figure dimensions.
    
    Returns:
        Tuple of (fig, ax) matplotlib objects.
    """
    fig, ax = plt.subplots(figsize=figsize)
    
    # Plot main curve
    ax.plot(x, y, color="black", linewidth=1.6)
    
    # Plot transition points with vertical lines
    for tp in transition_points:
        ax.scatter(tp, np.sin(tp), color="black", s=30, zorder=5)
        ax.axvline(tp, color="black", linestyle=":", linewidth=1.6, alpha=0.6)
    
    # Add motif labels
    label_y = 0.9
    label_positions = [0.16 * np.pi, np.pi, 1.75 * np.pi]
    label_texts = ["Increasing", "Decreasing", "Increasing"]
    
    for xp, txt in zip(label_positions, label_texts):
        ax.text(xp, label_y, txt, ha="center", va="center", fontsize=20)
    
    # Format axes
    ax.set_xticks([0, np.pi / 2, np.pi, 3 * np.pi / 2, 2 * np.pi])
    ax.set_xticklabels([
        r"$0$", 
        r"$\frac{\pi}{2}$", 
        r"$\pi$", 
        r"$\frac{3\pi}{2}$", 
        r"$2\pi$"
    ])
    ax.set_yticks([-1, 0, 1])
    ax.set_xlabel("$x$")
    ax.set_ylabel(r"$\sin(x)$")
    ax.grid(False)
    
    return fig, ax
```

```python
# Composition visualisation

# Generate sine curve
x, y = create_sine_curve(n_points=400)

# Get transition points
transition_points = get_transition_points()

# Create plot
fig, ax = plot_motifs_composition(x, y, transition_points)

plt.show()
```
